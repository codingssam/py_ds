```python
data = pd.read_csv("macrodata.csv")
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year, 
                         quarter=data.quarter,
                         name="date")
columns = pd.Index(["realgdp","infl", "unemp"], name="item")
```

```python
data = data.reindex(columns=columns)
```

```python
data
```

```python
data.index
```

```python
data.index = periods.to_timestamp("D", "end")
```

```python
ldata = data.stack().reset_index().rename(columns={0: "value"})
ldata
```

```python
ldata.pivot(index="date", columns="item", values="value")
```

```python
ldata.pivot(index="date", columns="item")
```

```python
ldata.set_index(["date", "item"]).unstack()
```

```python
df = pd.DataFrame({
    "key": ["foo", "bar", "baz"],
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "C": [7, 8, 9]
})
df
```

```python
df.set_index("key", inplace=True)
```

```python
df.columns.name = "variable"
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df.columns.name = ""
df
```

```python
melted = pd.melt(df, id_vars=["key"], 
                 var_name="variable", value_name="value",
                 value_vars=["A", "B"])
melted
```

```python
pivoted = melted.pivot(index="key", columns="variable", 
                       values="value")
pivoted
```

```python
pivoted.reset_index()
```
